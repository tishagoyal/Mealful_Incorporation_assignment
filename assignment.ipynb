```python
item_date_list=pd.DataFrame(df.item_date.unique())
item_date_list
```

```python
final_dic={}
# test=df[(df["schedule_time"]>="12:00:00")&(df["schedule_time"]<"15:00:00")&(df['item_date']=="2021-05-19")]
# len(test)
time_slot_list=["12AM_3AM","3AM_6AM","6AM_9AM","9AM_12PM","12PM_3PM","3PM_6PM","6PM_9PM","9PM_12AM"]
dic_times_slot={}
for date in item_date_list[0]:
    test=df[(df["schedule_time"]>="00:00:00")&(df["schedule_time"]<"03:00:00")&(df['item_date']==date)]
    dic_times_slot[time_slot_list[0]]=len(test)
    test1=df[(df["schedule_time"]>="03:00:00")&(df["schedule_time"]<"06:00:00")&(df['item_date']==date)]
    dic_times_slot[time_slot_list[1]]=len(test1)
    test2=df[(df["schedule_time"]>="06:00:00")&(df["schedule_time"]<"09:00:00")&(df['item_date']==date)]
    dic_times_slot[time_slot_list[2]]=len(test2)
    test3=df[(df["schedule_time"]>="09:00:00")&(df["schedule_time"]<"12:00:00")&(df['item_date']==date)]
    dic_times_slot[time_slot_list[3]]=len(test3)
    test4=df[(df["schedule_time"]>="12:00:00")&(df["schedule_time"]<"15:00:00")&(df['item_date']==date)]
    dic_times_slot[time_slot_list[4]]=len(test4)
    test5=df[(df["schedule_time"]>="15:00:00")&(df["schedule_time"]<"18:00:00")&(df['item_date']==date)]
    dic_times_slot[time_slot_list[5]]=len(test5)
    test6=df[(df["schedule_time"]>="18:00:00")&(df["schedule_time"]<"21:00:00")&(df['item_date']==date)]
    dic_times_slot[time_slot_list[6]]=len(test6)
    test7=df[(df["schedule_time"]>="21:00:00")&(df["schedule_time"]<"00:00:00")&(df['item_date']==date)]
    dic_times_slot[time_slot_list[7]]=len(test7)
    dic_times_slot
    
    final_dic[date]=dic_times_slot
```
